```python
import seaborn as sns
import pandas as pd

dataset = sns.load_dataset("titanic")
df = pd.DataFrame(dataset)
```

```python
print("1. Dataset Structure Information")
print("\n1.1 Number of rows and columns:\n", df.shape)
```

```python
print("\n1.2 Data types, nulls:\n", df.info())
```

```python
print("1.3 First 5 rows:")
df.head()
```

1.4 Summary Statistics

```python
df.describe()
```

2. Missing Value Analysis

```python
print("Null entries:\n", df.isnull().sum())
print()
print("Duplicate entries:", df.duplicated().sum())
```

3. Correlation Analysis

```python
new_df = df[['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare']].copy()
sns.heatmap(new_df.corr(), annot=True, cmap="coolwarm")
```

4. Plots

```python
sns.pairplot(df[['age', 'fare', 'survived']], hue='survived')
```

```python
pd.crosstab(df['pclass'], df['survived'])
sns.countplot(x='pclass', hue='survived', data=df)
```

```python
sns.boxplot(x="survived", y="age", data=df)
```

```python
sns.scatterplot(x='age', y='fare', data=df)
```
