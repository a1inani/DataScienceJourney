## Practice

- Plot histogram of Titanic ages.
- Create bar chart of survival rates by gender.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset = sns.load_dataset('titanic')
df = pd.DataFrame(dataset)
df.drop_duplicates(inplace=True)

# Plot histogram of Titanic ages
plt.title("Histogram of Titanic ages")
df['age'].plot(kind = 'hist')
plt.show()
```

```python
# Create a bar chart of survival rates by gender

male = 0
female = 0
for x in df.index:
    if df.loc[x, "sex"] == "male" and df.loc[x, 'survived'] == 1:
        male += 1
    elif df.loc[x, "sex"] == "female" and df.loc[x, 'survived'] == 1:
        female += 1

data = {'Gender': ['Male', 'Female'],
        'Survivors': [male, female]}
df2 = pd.DataFrame(data)

df2.plot(x='Gender', y='Survivors', kind='bar', title="Survival Rates by Gender")
plt.show()
```
