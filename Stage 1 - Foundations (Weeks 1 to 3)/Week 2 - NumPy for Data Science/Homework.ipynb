- Simulate 10,000 dice rolls
- Plot histogram of outcomes
- Compare observed probabilities with theoretical probabilities

```python
# Simulate 10,000 dice rolls

import numpy as np

dice_rolls = np.random.randint(1, 7, size=(10000))
print(dice_rolls)
```

```python
# Plot histogram of outcomes
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(dice_rolls)
plt.title("Random Dice Rolls (10,000 rolls)")
plt.show()
```

```python
# Compare observed probabilities with theoretical probabilities.

import pandas as pd

# count occurrences
counts = pd.Series(dice_rolls).value_counts().sort_index()

observed_probs = counts / len(dice_rolls)

theoretical_probs = pd.Series([1/6]*6, index=[1, 2, 3, 4, 5, 6])

comparison = pd.DataFrame({
    "Observed": observed_probs,
    "Theoretical": theoretical_probs
})
print(comparison)

comparison.plot(kind="bar")
plt.title("Observed vs Theoretical Dice Probabilities")
plt.ylabel("Probability")
plt.show()
```
