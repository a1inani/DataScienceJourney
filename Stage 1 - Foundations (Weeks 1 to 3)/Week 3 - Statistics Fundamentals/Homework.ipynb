# Homework

- Generate a dataset of 100 exam scores (mean=70, sd=10)
- Compute mean, std dev, 95% confidence interval and plot histogram.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# generate dataset

dataset = np.random.normal(loc=70, scale=10, size=100)

# compute mean
mean = np.mean(dataset)
print(f"Mean: {mean:.2f}")

# compute std dev
std_dev = np.std(dataset)
print(f"Std dev: {std_dev:.2f}")

# compute 95% confidence interval
confidence_level = 0.95
df = len(dataset)
conf_interval = stats.t.interval(confidence_level, df, loc=mean, scale=std_dev/np.sqrt(len(dataset)))
print(f"The 95% confidence interval is: {conf_interval[0]:.2f}, {conf_interval[1]:.2f}")

# plot histogram
plt.hist(dataset, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')
plt.xlabel("Score")
plt.ylabel("Probability Density")
plt.title("Histogram")
plt.show()
```
